```python
x = np.array(df['Text'])
y = np.array(df['language'])
```

```python
type(x)
```

```python
y
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
x_cv = cv.fit_transform(x)
print(x_cv)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x_cv, y, test_size= 0.2, random_state= 42)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
mb = MultinomialNB()
mb.fit(xtrain, ytrain)
```

```python
mb.score(xtest, ytest)
```

```python
user = "효모는 세포 수준의 생물학에서 모델 생물의 첫 번째 본보기로 간주해도 좋을 것이다 호흡 및 효소에 대한 생화학 분야는 여기에서 시작된다 이어 아카빤카비가 유전학 연구에서 이용되면서 유명해졌다 그 이후에도 기초 생물학의 여러 분야는 미생물을 모델로 사용하여 왔으나 오히려 균류보다 더 작고 단순한 미생물이다 박테리아류나 바이러스가 분자 생물학이나 유전학의 모델로 사용되는 것이 와 분열 효모 schizosaccharomyces pombe이다"
data = cv.transform([user]).toarray()
pred = mb.predict(data)
print(pred)
```
